```python
# web crawling using bing image downloader
import os
import shutil
import time
import random
from bing_image_downloader import downloader

# --- CONFIGURATION ---
base_dir = "raw_images"
LIMIT_PER_QUERY = 50 
TARGET_BIRD = "Scarlet Macaw" 

# Ensure base directory exists
os.makedirs(base_dir, exist_ok=True)

for bird in bird_data:
    bird_class = bird["class"]
    
    if TARGET_BIRD != "All" and TARGET_BIRD != bird_class:
        continue

    # Create the unified class folder first
    target_folder = os.path.join(base_dir, bird_class)
    os.makedirs(target_folder, exist_ok=True)
    
    print(f"\n{'='*50}\nBING SEARCHING FOR: {bird_class.upper()}\n{'='*50}")

    for q in bird["queries"]:
        clean_q = q.replace(" ", "_").lower()[:12]
        print(f"\n--- Downloading: {q} ---")

        # FIX 1: Download into a clearly separate temp folder to avoid path collisions
        downloader.download(q, limit=LIMIT_PER_QUERY, output_dir="temp_download", 
                            force_replace=False, verbose=False, timeout=60)
        
        temp_query_folder = os.path.join("temp_download", q)
        
        if os.path.exists(temp_query_folder):
            for i, f in enumerate(os.listdir(temp_query_folder)):
                # FIX 2: Skip potential hidden files or empty directories
                src_path = os.path.join(temp_query_folder, f)
                if not os.path.isfile(src_path):
                    continue

                ext = os.path.splitext(f)[1] or ".jpg"
                new_filename = f"bing_{clean_q}_{i:03d}{ext}"
                dest_path = os.path.join(target_folder, new_filename)
                
                # FIX 3: Use a try-except block for the move operation
                try:
                    if not os.path.exists(dest_path):
                        shutil.move(src_path, dest_path)
                except FileNotFoundError:
                    print(f"Skipping {f}: File disappeared during processing.")
            
            # Clean up the temp folder for this query
            try:
                shutil.rmtree("temp_download")
            except:
                pass

        wait_time = random.uniform(2, 5)
        print(f"Waiting {wait_time:.1f}s...")
        time.sleep(wait_time)

print(f"\nDone! All images are unified in '{base_dir}/Bird Name/'")
```

```python
# verify file format (make sure they are image files)
import os
from PIL import Image
from tqdm import tqdm

base_dir = "raw_images"
min_size = (10, 10)  # Minimum width and height
removed_count = 0

print(f"Starting validation in {base_dir}...")

# Iterate through every bird class folder
for root, dirs, files in os.walk(base_dir):
    for filename in tqdm(files, desc=f"Cleaning {os.path.basename(root)}"):
        file_path = os.path.join(root, filename)
        
        try:
            # 1. Check if the file is actually an image and not corrupted
            with Image.open(file_path) as img:
                img.verify()  # Verify that it's a valid image
            
            # 2. Re-open to check dimensions (verify() closes the file)
            with Image.open(file_path) as img:
                width, height = img.size
                
                # 3. Remove tiny images/icons/logos
                if width < min_size[0] or height < min_size[1]:
                    os.remove(file_path)
                    removed_count += 1
                    
        except (IOError, SyntaxError, Image.UnidentifiedImageError):
            # 4. Delete corrupted files or non-image files (like HTML)
            os.remove(file_path)
            removed_count += 1
        except Exception as e:
            print(f"Error processing {file_path}: {e}")

print(f"\nCleanup complete!")
print(f"Total corrupted or invalid images removed: {removed_count}")
```

```python
# filter image (remove no color )
import os
from PIL import Image, ImageStat
from tqdm import tqdm

base_dir = "raw_images"
removed_bw = 0
removed_simple = 0

def is_grayscale(img):
    """Returns True if the image is black and white or grayscale."""
    if img.mode == 'L':
        return True
    if img.mode == 'RGB':
        # Check if R, G, and B channels are almost identical
        stat = ImageStat.Stat(img)
        if sum(stat.var) < 30:  # Very low variance between channels means no color
            return True
    return False

print("Filtering out non-color and simple images...")

for root, dirs, files in os.walk(base_dir):
    for filename in tqdm(files, desc=f"Checking {os.path.basename(root)}"):
        file_path = os.path.join(root, filename)
        try:
            with Image.open(file_path) as img:
                img = img.convert("RGB")
                
                # 1. Remove Black and White
                if is_grayscale(img):
                    os.remove(file_path)
                    removed_bw += 1
                    continue
                
                # 2. Remove "Simple" images (Clip art / Cartoons often have very few colors)
                # We check the number of unique colors; if it's extremely low, it's likely a graphic
                colors = img.getcolors(maxcolors=1000)
                if colors is not None and len(colors) < 250:
                    os.remove(file_path)
                    removed_simple += 1
                    
        except:
            os.remove(file_path)

print(f"\nCleaned up!")
print(f"Removed B&W: {removed_bw}")
print(f"Removed Simple/ClipArt: {removed_simple}")
```

```python
# filter image (animated image)
import os
import torch
import shutil
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
from tqdm import tqdm

# 1. Setup - Force CPU usage
device = torch.device("cpu")
model = models.resnet50(weights='IMAGENET1K_V1').to(device)
model.eval()

# ImageNet IDs for various birds (broad spectrum)
BIRD_INDICES = list(range(7, 25)) + list(range(80, 100))

# Standard Preprocessing
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

base_dir = "raw_images"
review_dir = "suspicious_images"
os.makedirs(review_dir, exist_ok=True)

print("Starting AI Bird Filter (CPU Mode)...")

for bird_class in os.listdir(base_dir):
    class_path = os.path.join(base_dir, bird_class)
    if not os.path.isdir(class_path): continue
    
    # Create subfolder in review_dir
    os.makedirs(os.path.join(review_dir, bird_class), exist_ok=True)
    
    files = [f for f in os.listdir(class_path)]
    
    for filename in tqdm(files, desc=f"Checking {bird_class}"):
        img_path = os.path.join(class_path, filename)
        
        try:
            with Image.open(img_path).convert('RGB') as img:
                img_t = transform(img).unsqueeze(0).to(device)
                
                with torch.no_grad():
                    output = model(img_t)
                    _, predicted = torch.max(output, 1)
                
                # If the AI doesn't think it's a bird, move it to review
                if predicted.item() not in BIRD_INDICES:
                    shutil.move(img_path, os.path.join(review_dir, bird_class, filename))
        except:
            # If the file is so broken the AI can't even open it, move it to review
            shutil.move(img_path, os.path.join(review_dir, bird_class, filename))

print(f"\nAI Filter Complete! Check the '{review_dir}' folder to see what was flagged.")
```

```python
# rename file
import os
from PIL import Image
from tqdm import tqdm

base_dir = "raw_images"
target_size = (300, 300)

print(f"Standardizing images in {base_dir}...")

# Iterate through each bird class folder
for bird_folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, bird_folder)
    
    if not os.path.isdir(folder_path):
        continue
        
    # Get all files in the folder
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    
    # Sort files to ensure consistent numbering if you run this multiple times
    files.sort()
    
    # We use a secondary list to store original paths so we don't 
    # accidentally process a file we just renamed.
    for index, filename in enumerate(tqdm(files, desc=f"Processing {bird_folder}")):
        old_path = os.path.join(folder_path, filename)
        
        # Clean the class name for the filename (remove spaces)
        clean_class_name = bird_folder.replace(" ", "")
        new_filename = f"{clean_class_name}{index + 1}.jpg"
        new_path = os.path.join(folder_path, new_filename)
        
        try:
            with Image.open(old_path) as img:
                # 1. Convert to RGB (removes Alpha channel from PNGs and fixes grayscale)
                img = img.convert("RGB")
                
                # 2. Resize to 300x300
                img = img.resize(target_size, Image.Resampling.LANCZOS)
                
                # 3. Save as the new filename
                img.save(new_path, "JPEG", quality=95)
            
            # 4. Remove the old file if it had a different name
            if old_path != new_path:
                os.remove(old_path)
                
        except Exception as e:
            print(f"Could not process {old_path}: {e}")

print("\nSuccess! All images are now renamed and resized.")
```
